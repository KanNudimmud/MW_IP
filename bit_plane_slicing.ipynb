```python
## Bit-Plane Slacing
import cv2
import matplotlib.pyplot as plt

# Load image
lynx = cv2.imread('lynx.jpg')

# Convert to grayscale
lynx = cv2.cvtColor(lynx,cv2.COLOR_BGR2GRAY)

# Display the image
plt.imshow(lynx,cmap='gray')
```

```python
# Slicing
import numpy as np

b = np.zeros(lynx.shape);
for i in range(8):
  b = (lynx >> i) & 1
  plt.subplots(figsize=(20,20))
  plt.subplot(2,4,i+1)
  plt.imshow(b,cmap='gray')
```
